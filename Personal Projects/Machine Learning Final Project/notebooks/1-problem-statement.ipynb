##### **PROBLEM STATEMENT**

# Context
As a planet that is quite close to the asteroids belt, there is a probability that the Earth can be hit by an asteroid, even it is small. Astronauts nowadays are keen to determine whether an asteroid can be hazardous. But in order to accomplish that, a long process and more scientist are required.

# Business Objective
The objective of this project is to predict whether an asteroid is potentially harmful to Earth. The main business benefit is to save astronauts large amount of time and energy to make predictions. 

# Output
The output of this project is either an asteroid is hazardous or not.

# Machine Learning Task
This project uses supervised learning as the type of learning because the target variable is already labeled and known. The type of supervised learning will be classification since the type of dependent variable is categorical.

# Performance Measure
We use several metrics to evaluate our model performance and accuracy, as follows.
* Accuracy
* Precision
* Recall
* F1-Score
